```python
from pythonpic.classes import Grid, Species, Simulation
%matplotlib inline
from pythonpic.visualization.time_snapshots import SpatialDistributionPlot
import matplotlib.pyplot as plt
```

```python
g = Grid(1, 1, 150)
s = Species(1, 1, 10000, g)
# s.distribute_uniformly(g.L)
sim = Simulation(g, [s])
```

```python
fig, ax = plt.subplots()
for profile, description in [
    ('linear', 'liniowa'),
    ('quadratic', 'kwadratowa'),
    ('exponential', 'wykladnicza')]:
    s.distribute_nonuniformly(1, 0.2, 0.2, 0.4, profile=profile)
    ax.hist(s.x,bins=150, histtype='step', label=description);
ax.set_xlabel("Położenie $x$")
ax.set_ylabel("Liczba cząstek")
ax.set_title("Przykład położeń cząstek dla różnych kszałtów preplazmy")
ax.set_xlim(0, g.L)
ax.legend()
fig.savefig('/home/dominik/preplasma.eps')

# SpatialDistributionPlot(sim, ax)
```
